```python
for index, row in df.iterrows():
    name = row['MenuItem']
    size = row['Size']
    category = row["Category"]
    dietary = row["Dietary"]
    calories = row["Calories"]
    protein = row["Protein"]
    fats = row["Fats"]
    saturated = row["Saturated Fat"]
    trans = row["Trans fat"]
    cholesterols = row["Cholesterols"]
    carbs = row["Carbs"]
    sugar = row["Total sugars"]
    sodium = row["Sodium"]
    
    nutrition = {
        "calories":calories,
        "protein":protein,
        "fats":fats,
        "saturated":saturated,
        "trans":trans,
        "cholesterol":cholesterols,
        "carbs":carbs,
        "sugar":sugar,
        "sodium":sodium
    }
    
    item = {
        "name":name,
        "company":"McDonald's",
        "size":size,
        "category":category,
        "dietary":dietary,
        "nutrition":nutrition
    }
    collection_name.insert_one(item)
```
